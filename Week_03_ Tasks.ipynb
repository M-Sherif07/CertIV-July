You are working as a junior embedded systems developer tasked with designing a basic LED control system for a small electronics workshop.

## Task -4  Multi-Blink Confirmation

You are tasked with building a confirmation signal system for a smart device prototype using a Raspberry Pi Pico.

The system works like this:

* A **button is pressed** to trigger the confirmation.

* Instead of blinking once, the LED will blink **three times** in a row (short ON/OFF cycles) to act as a **confirmation signal**.

* Each time the button is pressed, the LED repeats the **multi-blink pattern**.

This behaviour is commonly used in the real world:

* **Remote control pairing**: a light blinks multiple times to confirm pairing success.

* **Security systems**: a lamp blinks to confirm an alarm has been armed/disarmed.

* **IoT devices**: LEDs blink in a pattern to indicate a successful Wi-Fi connection.

To practice structured coding, you are required to write the **multi-blink behaviour inside a function**. The main loop should only call the function when the button is pressed.

### Hardware Requirements
* Raspberry Pi Pico W
* LED light
* One Button
* Breadboard
* Jumper wires
