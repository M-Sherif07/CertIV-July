# Week_9_Assessment_Task


## Scenario

You have been commissioned by the **National Technology Museum** to prototype an **IoT-based motion detection alarm system** for its artifact storage room.
The museum wants a **simple, reliable, and interactive system** that alerts staff to any motion detected after hours, while allowing authorised users to arm, disarm, and mute the system using **physical buttons**.

Your prototype will be deployed on a **Raspberry Pi Pico** microcontroller and must integrate sensors, actuators, and status outputs to mimic a real-world smart security device.

##  Project Objective

Design, build, and test a working **Museum Heist Detector** that:

1. Uses a **PIR motion sensor** to detect movement.
2. Activates a **buzzer and LED** alarm when motion is detected.
3. Uses **three physical buttons** to:

   * Arm/Disarm the system.
   * Mute/Unmute the buzzer.
   * Display system status on the serial monitor.

## Software Requirements

* Use **MicroPython** language.
* Code must be **commented** to explain each function and logic section.
* Include appropriate **debounce handling** for button presses.
* Ensure the program exits safely on `KeyboardInterrupt`.

###  Assessment Criteria
- Circuit correctly wired according to pin configuration
- Code runs without syntax or runtime errors
- PIR sensor successfully detects motion
- LED and buzzer activate when motion is detected and the system is armed
- Arm/Disarm button toggles system correctly
- Mute/Unmute button correctly silences buzzer
- Status button prints the current system state to the terminal
- Code includes meaningful comments and clear structure
- Student explains system behaviour in written reflection
